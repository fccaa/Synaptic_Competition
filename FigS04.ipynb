```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def E_func(w0, w1):
    w_diff = w0-w1
    E = -0.25 * np.sum(np.square(w_diff)) * 0.5 
    
    tmp = 0
    for i in range(0,len(w_diff), 1):
        for j in range(i+1,len(w_diff), 1):
            tmp += w_diff[i] * w_diff[j] * 0.5
            
    E -= 0.5 * tmp
    
    return E
        
```

```python
w1 = np.random.rand(100)
w1 *= 1 / w1.mean()
```

```python
w1.mean()
```

```python
w2 = np.random.rand(len(w1))
w2 *= 0 / w2.mean()
E_func(w1, w2)
```

```python
x = np.arange(0,1.000002,0.02)
x1 = np.arange(0,1.000002,0.2)
```

```python
y_all = []


for xi0 in x1:
    y = []

    for xi1 in x:
        w1 = np.random.rand(100)
        w1 *= 1 / w1.mean()
        w2 = np.random.rand(len(w1))
        w2 *= xi1 / w2.mean()
        y.append(E_func(w1*xi0, w2))
        
    y_all.append(y)
```

```python

papersize=(11.4/2.525, 8.4/2.525)
fig, ax = plt.subplots(1,1,figsize=papersize)

params = {'legend.handlelength': 4}
plt.rcParams.update(params)
plt.rcParams.update({'font.size': 8})
# plt.rcParams.update({'font.family': "Sans"})
plt.rcParams.update({'mathtext.fontset': "stix"})

ax.plot(x,y_all[0], linestyle = 'solid', label=r'$\langle w_{1j} \rangle_j =0$')
ax.plot(x,y_all[1], linestyle = 'dotted', lw = 2, label=r'$\langle w_{1j} \rangle_j =0.2$')
ax.plot(x,y_all[2], linestyle = (0, (3,5,1,5)), label=r'$\langle w_{1j} \rangle_j =0.4$')
ax.plot(x,y_all[3], linestyle = 'dashdot', label=r'$\langle w_{1j} \rangle_j =0.6$')
ax.plot(x,y_all[4], linestyle = (0, (10,3)), label=r'$\langle w_{1j} \rangle_j =0.8$')
ax.plot(x,y_all[5], linestyle = 'dashed', label=r'$\langle w_{1j} \rangle_j =1.0$')

ax.set_xlim((0,1))
ax.legend()

ax.set_xlabel(r'Average Magnitude of $w_{2j}$, $\langle w_{2j} \rangle_j$')
ax.set_ylabel("Analogous Energy")

plt.tight_layout()
plt.savefig("FigS03.pdf")
```
