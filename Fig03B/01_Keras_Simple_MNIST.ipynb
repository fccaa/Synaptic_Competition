```python
# Import modules

import numpy
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.datasets import mnist

```

```python
# Download the data and put it in numpy array

(x_train_ori, y_train_ori), (x_test_ori, y_test_ori) = mnist.load_data()
```

```python
results = []

N_test = 5000

for N in range(100, 1001, 100):
    
    result = []
    
    for t in range(10):
        
        sample_idx = numpy.random.choice(numpy.arange(x_train_ori.shape[0]), 
                                        size=N_test, replace=False)
    
        # Reshape the input patterns from 2D to 1D

        x_train = x_train_ori.reshape(x_train_ori.shape[0], x_train_ori.shape[1]*x_train_ori.shape[2])[sample_idx]
        x_test = x_test_ori.reshape(x_test_ori.shape[0], x_test_ori.shape[1]*x_test_ori.shape[2])

        # Change from int to float32

        x_train = x_train.astype('float32')
        x_test = x_test.astype('float32')

        # Convert it the less than 1.0

        x_train /= 255.0
        x_test  /= 255.0

        # Original y are [0, 1, 2, ..., 9]
        # Convert it to (1,0,...), (0,1,...), etc

        y_train = keras.utils.to_categorical(y_train_ori, 10)[sample_idx]
        y_test = keras.utils.to_categorical(y_test_ori, 10)

        # Construct the network
        # input nodes -> 100 nodes -> 10 output nodes

        model = Sequential()
        model.add(Dense(N, activation="relu", input_dim=x_train.shape[1]))
        model.add(Dense(10, activation="softmax"))# Compile (finalize) the model

        model.compile(loss='categorical_crossentropy',
                  optimizer='sgd',
                  metrics=['accuracy'])

        model.fit(x_train, y_train, epochs=5, verbose=0)

        result.append(model.evaluate(x_test, y_test, verbose=0)[1])
        
    print(N, numpy.mean(result), sep="\t")
    results.append(result)
```
